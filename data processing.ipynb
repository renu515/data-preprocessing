```python
# -------------------------------
# Problem 1: Employee Bonus Eligibility
# -------------------------------

staff_scores = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

all_scores = list(staff_scores.values())
best_score = max(all_scores)

bonus_candidates = []

for emp_name, emp_score in staff_scores.items():
    # Checking who matches the highest score
    if emp_score == best_score:
        bonus_candidates.append(emp_name)

result_line = ", ".join(bonus_candidates)

print("Top Performers Eligible for Bonus:",
      result_line,
      "(Score:", best_score, ")")


# -------------------------------
# Problem 2: Search Query Keyword Analysis
# -------------------------------

raw_query = "Buy mobile phone buy phone online"
query_text = raw_query.lower()   # normalize to lowercase

import string  # keeping import here since that's how it was written

# remove punctuation step by step
for symbol in string.punctuation:
    query_text = query_text.replace(symbol, "")

split_words = query_text.split()

word_counter = {}

for w in split_words:
    current_count = word_counter.get(w, 0)
    word_counter[w] = current_count + 1

# find duplicates manually
duplicates = {}

for key, val in word_counter.items():
    if val > 1:
        duplicates[key] = val

print("\nRepeated Keywords:", duplicates)


# -------------------------------
# Problem 3: Sensor Data Validation
# -------------------------------

sensor_values = [3, 4, 7, 8, 10, 12, 5]
even_readings = []

for idx in range(len(sensor_values)):   # using range intentionally
    reading = sensor_values[idx]

    # store only even readings
    if reading % 2 == 0:
        even_readings.append((idx, reading))

print("\nValid Sensor Readings:", even_readings)


# -------------------------------
# Problem 4: Email Domain Usage Analysis
# -------------------------------

email_list = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_stats = {}

for e in email_list:
    parts = e.split("@")

    # small safety check
    if len(parts) > 1:
        domain_name = parts[1]
    else:
        domain_name = ""

    if domain_name not in domain_stats:
        domain_stats[domain_name] = 1
    else:
        domain_stats[domain_name] += 1

total_emails = len(email_list)

print("\nEmail Domain Usage:")

for d, c in domain_stats.items():
    percent = (c / total_emails) * 100
    rounded_percent = round(percent)
    print(d + ":", str(rounded_percent) + "%")


# -------------------------------
# Problem 5: Sales Spike Detection
# -------------------------------

daily_sales = [1200, 1500, 900, 2200, 1400, 3000]

avg_sale_value = sum(daily_sales) / len(daily_sales)
spike_threshold = avg_sale_value * 1.3   # 30% above average

print("\nSales Spike Days:")

for i in range(len(daily_sales)):
    sale_amount = daily_sales[i]

    if sale_amount > spike_threshold:
        print("Day", i + 1, ":", sale_amount)


# -------------------------------
# Problem 6: Duplicate User ID Detection
# -------------------------------

users = ["user1", "user2", "user1", "user3", "user1", "user3"]

user_frequency = {}

for user in users:
    if user in user_frequency:
        user_frequency[user] += 1
    else:
        user_frequency[user] = 1

print("\nDuplicate User IDs:")

for user, freq in user_frequency.items():
    if freq > 1:
        print(user, "->", freq, "times")
```
